# Aumento das vendas 

```python
# mostra o quanto a empresa vendeu nos meses seguintes 
dados['aumento'] = dados['vendas'].diff() # faz conta de menos
                                    # 20 - 10 = 10
                                    #33 - 10 = 10
                                    
dados.head()
```

```python
# cor do grafico
sns.set_palette('Accent')
sns.set_style('darkgrid')

# selecionando grafico
ax = sns.lineplot(x='mes', y='aumento', data=dados)

# tamanho do grafico
ax.figure.set_size_inches(12,6)

                                             # titulo vai fica posicionado na esquerda
ax.set_title('Aumento das Vendas da alucar 2017 e 2018', loc = 'left', fontsize = 18)

ax.set_xlabel('Tempo', fontsize = 14)

ax.set_ylabel('Aumento', fontsize = 14)

ax = ax
```

```python
dados['Aceleração'] = dados['aumento'].diff()
dados.head()
```

# Analisando a sazonalidade

```python
                                                       # mostra o dia da semana
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.day_name()
vendas_por_dia.head()
```

```python
vendas_por_dia['dia_da_semana'].unique()
```

```python
dias_traduzidos = {'Monday': 'Segunda', 'Tuesday': 'Terça', 'Wednesday': 'Quarta', 'Thursday' : 'Quinta', 'Friday': 'Sexta', 'Saturday': 'Sabado',
       'Sunday': 'Domingo'}
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map(dias_traduzidos)
```

```python
vendas_por_dia.head()
```

# Agrupando os dias 

```python
vendas_agrupadas = vendas_por_dia.groupby('dia_da_semana')['vendas', 'aumento', 'Aceleração'].mean().round(2)
vendas_agrupadas
```

# Statsmodels

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose([chocolate['vendas']], freq=2)
ax = resultado.plot()
```

```python
observação = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = ({
    'observação': observação,
    'tendencia': tendencia,
    'sazonalidade': sazonalidade,
    'ruido': ruido
})
resultado = pd.DataFrame(data)
resultado.head()
```

```python
plt.figure(figsize =(16,12))


# 3 = dividir em 3 partes os graficos 
# 1 = 1 ocupe todo o espaço
# 1 = 1 seja a primeira parte
ax = plt.subplot(3, 1, 1)

ax.set_title('Exemplos de statsmodels', fontsize=18, loc = 'left')

# selecionando grafico
ax = sns.lineplot(x=resultado.index,  y='observação', data=resultado)
plt.subplot(3, 1, 2)

# selecionando grafico
ax = sns.lineplot(x=resultado.index, y='tendencia', data=resultado)
plt.subplot(3,1,3)

# selecionando grafico
ax = sns.lineplot(x=resultado.index, y='ruido', data=resultado)
```
